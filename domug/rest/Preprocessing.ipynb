```python
import pandas as pd
import numpy as np
import re

# TBD 채워 넣을 기준이 되는 데이터프레임
df_3 = pd.read_csv("challenger_3min.csv")

df_3
```

```python
# 챔피언 이름 리스트 자료형으로 바꾸는 함수
def champ_name_to_list(df):
    for i in range(len(df)):
        df.at[i, "blueChamps"] = re.findall("\d+", df.blueChamps[i])
        df.at[i, "redChamps"] = re.findall("\d+", df.redChamps[i])
```

```python
champ_name_to_list(df_3)
```

```python
# 3분 데이터를 기준으로 TBD 채우는 함수
def preprocess(df, minute):
    data = pd.read_csv("{}mins/challenger_{}min.csv".format(minute, minute))
    
    tally = 0
    for i in range(len(df)):
        gameid = df.iloc[i, 0]
        if data[data.gameId == gameid] is not None:
            try:
                index = data[data.gameId == gameid].index[0]
                data.at[index, "blueWins"] = df.at[i, "blueWins"]
                data.at[index, "blueChamps"] = df.at[i, "blueChamps"]
                data.at[index, "redWins"] = df.at[i, "redWins"]
                data.at[index, "redChamps"] = df.at[i, "redChamps"]
                tally += 1

            except:
                pass

        if tally % 4000 == 0:
            print("{}개 진행완료".format(tally))

    
    data = data.replace("TBD", np.nan)
    data.dropna(inplace=True)
    data.to_csv("{}mins/challenger_{}min_modified.csv".format(minute, minute), index=False)
```

```python
id1 = df1.gameId.unique()
common_ids = id1.tolist()
```

```python
# df1 ~ df18까지 전부해줘야함
id2 = df18.gameId.unique()
for ids in common_ids:
    if ids not in id2:
        common_ids.remove(ids)
        
len(common_ids)
```

```python
# 모든 timestamp에서 공통된 gameid 이외의 행 제거
def delete_rows(df, name):
    remove_index = []
    for i in range(len(df)):
        if df.gameId[i] not in common_ids:
            remove_index.append(i)
    df.drop(remove_index, axis=0, inplace=True)
    df.reset_index().drop("index", axis=1)
    df.to_csv("Final Dataset/{}.csv".format(name), index=False)
```
