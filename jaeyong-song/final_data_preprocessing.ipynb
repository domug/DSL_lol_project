```python
df_30 = df_30.drop(['redWins'], axis=1)
df_30 = df_30.replace({'blueWins': 'Win'}, {'blueWins': 1})
df_30 = df_30.replace({'blueWins': 'Fail'}, {'blueWins': 0})
df_30
```

```python
df_30 = df_30.replace({'blueFirstTowerLane': "[\'TOP_LANE\']"}, {'blueFirstTowerLane': 1})
df_30 = df_30.replace({'blueFirstTowerLane': "[\'MID_LANE\']"}, {'blueFirstTowerLane': 2})
df_30 = df_30.replace({'blueFirstTowerLane': "[\'BOT_LANE\']"}, {'blueFirstTowerLane': 3})
df_30 = df_30.replace({'blueFirstTowerLane': "[]"}, {'blueFirstTowerLane': 0})
df_30[['blueFirstTowerLane']]
```

```python
df_30 = df_30.replace({'redFirstTowerLane': "[\'TOP_LANE\']"}, {'redFirstTowerLane': 1})
df_30 = df_30.replace({'redFirstTowerLane': "[\'MID_LANE\']"}, {'redFirstTowerLane': 2})
df_30 = df_30.replace({'redFirstTowerLane': "[\'BOT_LANE\']"}, {'redFirstTowerLane': 3})
df_30 = df_30.replace({'redFirstTowerLane': "[]"}, {'redFirstTowerLane': 0})
df_30[['redFirstTowerLane']]
```

'DragonType'에 대한 전처리<br>
blue와 red에 대해서 각 드래곤 숫자로 변환해서 표현하면 됨.<br>


```python
df_30[['blueDragonType']]
```

```python
df_30['blueAirDragon'] = 0
df_30['blueEarthDragon'] = 0
df_30['blueWaterDragon'] = 0
df_30['blueFireDragon'] = 0
df_30['blueElderDragon'] = 0
df_30['redAirDragon'] = 0
df_30['redEarthDragon'] = 0
df_30['redWaterDragon'] = 0
df_30['redFireDragon'] = 0
df_30['redElderDragon'] = 0
```

```python
def split_blue_dragon(df, col):
    for i in df.index:
        val = df.loc[i, col]
        splitted = val[1:-1].split(', ')
        for dragon in splitted:
            if(dragon == "'AIR_DRAGON'"):
                df.loc[i, 'blueAirDragon'] = df.loc[i, 'blueAirDragon'] + 1
            elif(dragon == "'EARTH_DRAGON'"):
                df.loc[i, 'blueEarthDragon'] = df.loc[i, 'blueEarthDragon'] + 1
            elif(dragon == "'WATER_DRAGON'"):
                df.loc[i, 'blueWaterDragon'] = df.loc[i, 'blueWaterDragon'] + 1
            elif(dragon == "'FIRE_DRAGON'"):
                df.loc[i, 'blueFireDragon'] = df.loc[i, 'blueFireDragon'] + 1
            elif(dragon == "'ELDER_DRAGON'"):
                df.loc[i, 'blueElderDragon'] = df.loc[i, 'blueElderDragon'] + 1

split_blue_dragon(df_30, 'blueDragonType')
df_30 = df_30.drop(['blueDragonType'], axis=1)
df_30
```

```python
def split_red_dragon(df, col):
    for i in df.index:
        val = df.loc[i, col]
        splitted = val[1:-1].split(', ')
        for dragon in splitted:
            if(dragon == "'AIR_DRAGON'"):
                df.loc[i, 'redAirDragon'] = df.loc[i, 'redAirDragon'] + 1
            elif(dragon == "'EARTH_DRAGON'"):
                df.loc[i, 'redEarthDragon'] = df.loc[i, 'redEarthDragon'] + 1
            elif(dragon == "'WATER_DRAGON'"):
                df.loc[i, 'redWaterDragon'] = df.loc[i, 'redWaterDragon'] + 1
            elif(dragon == "'FIRE_DRAGON'"):
                df.loc[i, 'redFireDragon'] = df.loc[i, 'redFireDragon'] + 1
            elif(dragon == "'ELDER_DRAGON'"):
                df.loc[i, 'redElderDragon'] = df.loc[i, 'redElderDragon'] + 1

split_red_dragon(df_30, 'redDragonType')
df_30 = df_30.drop(['redDragonType'], axis=1)
df_30
```
